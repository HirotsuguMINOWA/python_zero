```python
import pickle
from collections import defaultdict
```

```python
with open('saizeriya.pickle','rb') as f:
    names, prices, cals = pickle.load(f)
```

```python
d = defaultdict(int)
```

```python
def search(n, p):
    if n < 0:
        return 0
    if d[(n,p)] is not 0:
        return d[(n,p)]
    
    c1 = 0
    if prices[n] <= p:
        c1 = cals[n] + search(n-1, p - prices[n])
    c2 = search(n-1, p)
    cmax = max(c1, c2)
    d[(n,p)] = cmax
    return cmax
```

```python
%%time
search(113, 10000)
```
